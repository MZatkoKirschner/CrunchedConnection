```python
#Import modules
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.ensemble import RandomForestClassifier
```

```python
#Read merged flight+met data for ML and perform slight processing
file = '../data/processed/merged/2019_FlightMetMerged.csv'
df = pd.read_csv(file)

df.drop_duplicates(inplace=True)

#For now remove flights where arrival time before departure time
df['ARR_TIME'] = pd.to_datetime(df['ARR_TIME'])
df['DEP_TIME'] = pd.to_datetime(df['DEP_TIME'])
df = df[df['ARR_TIME']>df['DEP_TIME']]
```

```python
#For df used for ML modeling 

#Drop any row from df with nans for ML model, drop columns columns that are not needed
dfML = df.dropna() 

#Merged arrival bins for simplicity for now
dfML['ARR_DELAY_GROUP'].loc[(dfML['ARR_DELAY_GROUP']<=0)] = 0
dfML['ARR_DELAY_GROUP'].loc[(dfML['ARR_DELAY_GROUP']>0)] = 1

#dfML['ARR_DELAY_GROUP'].loc[(dfML['ARR_DELAY_GROUP']<=0)] = 0
#dfML['ARR_DELAY_GROUP'].loc[(dfML['ARR_DELAY_GROUP']>0) & (dfML['ARR_DELAY_GROUP']<4)] = 1
#dfML['ARR_DELAY_GROUP'].loc[(dfML['ARR_DELAY_GROUP']>=4) & (dfML['ARR_DELAY_GROUP']<8)] = 2
#dfML['ARR_DELAY_GROUP'].loc[(dfML['ARR_DELAY_GROUP']>=8)] = 3
```

```python
#X = dfMLtest[['tmpF_D', 'dptF_D','spd_D', '6hPrecPrb_D', 
#          'fzRnPrb_D', 'snowPrb_D', '6hrTsPrb_15mi_D',
#          '6hrSvrTsPrb_25mi_D', 'tmpF_A', 'dptF_A', 'spd_A', 
#          'fzRnPrb_A', 'snowPrb_A', '6hrTsPrb_15mi_A','6hrSvrTsPrb_25mi_A',
#          '12hPrecPrb_D','12hPrecPrb_A','6hPrecPrb_A']]

X = dfML[['tmpF_D', 'spd_D', 'fzRnPrb_D', 'snowPrb_D', '6hrTsPrb_15mi_D',
              '6hrSvrTsPrb_25mi_D', 'tmpF_A', 'spd_A', 'fzRnPrb_A', 'snowPrb_A', 
              '6hrTsPrb_15mi_A','6hrSvrTsPrb_25mi_A','6hPrecPrb_D','6hPrecPrb_A','12hPrecPrb_D',
              '12hPrecPrb_A']]

y = dfML['ARR_DELAY_GROUP']
```

```python
#Experiment with sqrt values
X['spd_D_sqrt'] = np.sqrt(X['spd_D'])
X['spd_A_sqrt'] = np.sqrt(X['spd_A'])

X['6hPrecPrb_D_sqrt'] = np.sqrt(X['6hPrecPrb_D'])
X['6hPrecPrb_A_sqrt'] = np.sqrt(X['6hPrecPrb_A'])

X['fzRnPrb_D_sqrt'] = np.sqrt(X['fzRnPrb_D'])
X['fzRnPrb_A_sqrt'] = np.sqrt(X['fzRnPrb_A'])

X['snowPrb_D_sqrt'] = np.sqrt(X['snowPrb_D'])
X['snowPrb_A_sqrt'] = np.sqrt(X['snowPrb_A'])

X['6hrTsPrb_15mi_D_sqrt'] = np.sqrt(X['6hrTsPrb_15mi_D'])
X['6hrTsPrb_15mi_A_sqrt'] = np.sqrt(X['6hrTsPrb_15mi_A'])

X['6hrSvrTsPrb_25mi_D_sqrt'] = np.sqrt(X['6hrSvrTsPrb_25mi_D'])
X['6hrSvrTsPrb_25mi_A_sqrt'] = np.sqrt(X['6hrSvrTsPrb_25mi_A'])

X['12hPrecPrb_D_sqrt'] = np.sqrt(X['12hPrecPrb_D'])
X['12hPrecPrb_A_sqrt'] = np.sqrt(X['12hPrecPrb_A'])
```

```python
#Add in square root of variables too
X['spd_D_log'] = np.log(X['spd_D']+1)
X['spd_A_log'] = np.log(X['spd_A']+1)

X['6hPrecPrb_D_log'] = np.log(X['6hPrecPrb_D']+1)
X['6hPrecPrb_A_log'] = np.log(X['6hPrecPrb_A']+1)

X['fzRnPrb_D_log'] = np.log(X['fzRnPrb_D']+1)
X['fzRnPrb_A_log'] = np.log(X['fzRnPrb_A']+1)

X['snowPrb_D_log'] = np.log(X['snowPrb_D']+1)
X['snowPrb_A_log'] = np.log(X['snowPrb_A']+1)

X['6hrTsPrb_15mi_D_log'] = np.log(X['6hrTsPrb_15mi_D']+1)
X['6hrTsPrb_15mi_A_log'] = np.log(X['6hrTsPrb_15mi_A']+1)

X['6hrSvrTsPrb_25mi_D_log'] = np.log(X['6hrSvrTsPrb_25mi_D']+1)
X['6hrSvrTsPrb_25mi_A_log'] = np.log(X['6hrSvrTsPrb_25mi_A']+1)

X['12hPrecPrb_D_log'] = np.log(X['12hPrecPrb_D']+1)
X['12hPrecPrb_A_log'] = np.log(X['12hPrecPrb_A']+1)
```

```python
#Add in square of log
X['spd_D_logsq'] = (np.log(X['spd_D']+1))**2
X['spd_A_logsq'] = (np.log(X['spd_A']+1))**2

X['6hPrecPrb_D_logsq'] = (np.log(X['6hPrecPrb_D']+1))**2
X['6hPrecPrb_A_logsq'] = (np.log(X['6hPrecPrb_A']+1))**2

X['fzRnPrb_D_logsq'] = (np.log(X['fzRnPrb_D']+1))**2
X['fzRnPrb_A_logsq'] = (np.log(X['fzRnPrb_A']+1))**2

X['snowPrb_D_logsq'] = (np.log(X['snowPrb_D']+1))**2
X['snowPrb_A_logsq'] = (np.log(X['snowPrb_A']+1))**2

X['6hrTsPrb_15mi_D_logsq'] = (np.log(X['6hrTsPrb_15mi_D']+1))**2
X['6hrTsPrb_15mi_A_logsq'] = (np.log(X['6hrTsPrb_15mi_A']+1))**2

X['6hrSvrTsPrb_25mi_D_logsq'] = (np.log(X['6hrSvrTsPrb_25mi_D']+1))**2
X['6hrSvrTsPrb_25mi_A_logsq'] = (np.log(X['6hrSvrTsPrb_25mi_A']+1))**2

X['12hPrecPrb_D_logsq'] = (np.log(X['12hPrecPrb_D']+1))**2
X['12hPrecPrb_A_logsq'] = (np.log(X['12hPrecPrb_A']+1))**2
```

```python
#Split conversion dataset into train and test groups
X_train, X_test, y_train, y_test = train_test_split(X, y)
```

```python
Xy_train = X_train.copy()
Xy_train['ARR_DELAY_GROUP'] = y_train
```

```python
Xy_train1 = Xy_train[Xy_train['ARR_DELAY_GROUP']==1]
Xy_train0 = Xy_train[Xy_train['ARR_DELAY_GROUP']==0].sample(n=len(Xy_train1))
Xy_train_subsetTmp = [Xy_train1,Xy_train0]
Xy_train_subset = pd.concat(Xy_train_subsetTmp)

Xy_train_subset = Xy_train_subset.sample(frac=1)

X_train_subset = Xy_train_subset.iloc[:,:-1]
y_train_subset = Xy_train_subset.iloc[:,-1]
```

```python
#Train random forest model
#clf = RandomForestClassifier(n_estimators=100,n_jobs=4,class_weight='balanced').fit(X_train, y_train)
clf = RandomForestClassifier(n_estimators=10,n_jobs=5,class_weight='balanced').fit(X_train_subset, y_train_subset)
```

```python
#Run trained model on test set
forest_predicted = clf.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from sklearn.metrics import classification_report

print('Classification Report')
#print(classification_report(y_test, forest_predicted,target_names=['< 15 min', '15-1hr','1hr-2hr','>2 hr']))
print(classification_report(y_test, forest_predicted,target_names=['0', '1']))

print (' ')

print('Micro-averaged precision = {:.2f} (treat instances equally)'
      .format(precision_score(y_test, forest_predicted, average = 'micro')))
print('Macro-averaged precision = {:.2f} (treat classes equally)'
      .format(precision_score(y_test, forest_predicted, average = 'macro')))
print('Micro-averaged f1 = {:.2f} (treat instances equally)'
      .format(f1_score(y_test, forest_predicted, average = 'micro')))
print('Macro-averaged f1 = {:.2f} (treat classes equally)'
      .format(f1_score(y_test, forest_predicted, average = 'macro')))
```
